```python
import matplotlib.pyplot as plt
from sklearn import datasets

iris = datasets.load_iris()

x = iris.data
y = iris.target

xcls0 = x[y==0]
xcls1 = x[y==1]
xcls2 = x[y==2]
```

```python
plt.scatter(xcls1[:,0], xcls1[:,1], color='r')
plt.scatter(xcls0[:,0], xcls0[:,1], color='b')
plt.scatter(xcls2[:,0], xcls2[:,1], color='g')
plt.show()
```

```python
from sklearn.cluster import KMeans
import numpy as np

x = np.array([[1,2],[1,4],[1,0],[10,2],[10,4],[10,0]])
print(x.shape)
```

```python
kmeans = KMeans(n_clusters=2, random_state=0)

kmeans.fit(x)

x_tst = np.array([[0,0],[12,3]])

print(kmeans.cluster_centers_)
```

```python
x_tst = np.array([[0,0],[12,3]])
predVals = kmeans.predict(x_tst)
print(predVals)
```

```python
np.random.seed(5)
x = iris.data
y = iris.target

kmeans = KMeans(n_clusters=2)
kmeans.fit(x)
print(kmeans.cluster_centers_)
```

```python
from mpl_toolkits.mplot3d import Axes3D

def plotFigure(x, name, est, fignum, titles):
    fig = plt.figure(fignum,figsize=(4,3))
    ax = Axes3D(fig, rect=[0,0,0.95,1], elev=48, azim=134)
    est.fit(x)
    labels = est.labels_
    ax.scatter(x[:,3], x[:,0], x[:,2], c=labels.astype(np.float), edgecolors='k')
    ax.dist = 12
    fignum = fignum + 1
    return fignum

estimators = [('k_means_iris_8', KMeans(n_clusters=8)),
              ('k_means_iris_3', KMeans(n_clusters=3)),
              ('k_means_iris_3_ma_init', KMeans(n_clusters=3, n_init=1))]

fignum = 1
titles = ['8 grupos', '3 grupos', '3 grupos ruim']
for name, est in estimators:
    fignum = plotFigure(x, name, est, fignum, titles)
```

```python
inertias = []
# Determinate the number of class
# The lower the inertia, the better the grouping
# yn -> clustering center dimension n
# xn -> data dimentsion n
# Inertia = (y1 - x1)^2 + (y2 - x2)^2 + ... + (yn - xn)
k = range(1,15)
for e in k:
    km = KMeans(n_clusters=e)
    km = km.fit(x)
    inertias.append(km.inertia_)

plt.plot(k, inertias, 'bx-')
plt.xlabel('k')
plt.ylabel('inertias')
plt.title('Elbow Method for k')
plt.show()
```

```python
plt.plot(k, inertias, 'bx-')
plt.xlabel('k')
plt.ylabel('inertias')
plt.text(3,100, 'The best k', family='monospace', fontsize=10)
plt.title('Elbow Method for k')
plt.show()
```
