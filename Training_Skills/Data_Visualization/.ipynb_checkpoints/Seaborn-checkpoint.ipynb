## Controlling Ploted Figure Aesthetics

```python
def sinplot(flip = 1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i*0.5)*(7 - i)*flip)
```

```python
sinplot()
```

```python
sns.set_style('ticks', {'axes.grid': True,'xtick.direction':'in'})
sinplot()
sns.despine(left = True, bottom = True)
```
