```python
import numpy as np
import matplotlib.pyplot as plt
import math
%matplotlib inline
```

```python
x = [0.5, 2.2, 2.0]
y = [2.0, 2.5, 1.4]
```

```python
alpha = 0.01
```

```python
w0 = 0.1
w1 = 0.1
```

```python
def y_hat(x, w0, w1):
    return w0 + w1*x
```

```python
y_hat(1.5, w0, w1)
```

```python
def plot_line(x, y, w0, w1):
    x_values = [i for i in range(int(min(x)) -1, int(max(x)) + 2)]
    y_values = [y_hat(x, w0, w1) for x in x_values]
    plt.plot(x_values, y_values, 'r')
    plt.plot(x,y,'bo')
```

```python
plot_line(x,y,w0,w1)
```

```python
def MSE(x, y, w0, w1):
    cost = 0
    n = len(x)
    for i in range(0, len(x)):
        cost += (y_hat(x[i], w0, w1) - y[i])**2
    return cost/float(n)
```

```python
MSE(x, y, w0, w1)
```

```python
def gradient_descent_step(w0, w1, x, y, alpha):
    
    error_w0 = 0
    error_w1 = 0
    n = float(len(x))
    
    for i in range(0, len(x)):
        error_w0 += y_hat(x[i], w0, w1) - y[i]
        error_w1 += (y_hat(x[i], w0, w1) - y[i]) * x[i]
        
    new_w0 = w0 - alpha * (1/n) * error_w0
    new_w1 = w1 - alpha * (1/n) * error_w1
    
    return new_w0, new_w1
```

```python
w0, w1 = gradient_descent_step(w0, w1, x, y, alpha)
```

```python
print("w0={} w1={}".format(w0, w1))
```

```python
q_rep = 800
def gradient_descent(w0, w1, x, y, alpha, repetitions):
    cost = np.zeros(repetitions)
    for i in range(repetitions):
        w0, w1 = gradient_descent_step(w0, w1, x, y, alpha)
        cost[i] = MSE(x, y, w0, w1)
    return w0, w1, cost
```

```python
w0, w1, cost = gradient_descent(w0, w1, x, y, alpha, q_rep)
print(cost)
```

```python
fig, ax = plt.subplots()
ax.plot(np.arange(q_rep), cost, 'r')
ax.set_xlabel('Iterations')
ax.set_ylabel('Cost')
ax.set_title('MSE vs Interations')
```
