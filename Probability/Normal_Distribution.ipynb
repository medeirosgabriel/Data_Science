```python
# Imports

import random
from matplotlib import pyplot as plt
from collections import Counter as Counter
import math

# Binomial Variable

def bernoulli_trial(p):
    return 1 if random.random() < p else 0
    
def binomial (n, p):
    return sum(bernoulli_trial(p) for _ in range(n))

# Cumulative Distribution Function - Returns the probability that a random variable is less than x;

def normal_cdf(x, mu=0,sigma=1):
    return (1 + math.erf((x - mu) / math.sqrt(2) / sigma)) / 2

# Histogram - With one unit interval (Cumulative Distribution Function)

def make_hist(p, n, num_points):
    data = [binomial(n, p) for _ in range(num_points)]
    histogram = Counter(data)
    # Use a Bar chart to show the  sample current binomials
    plt.bar([x - 0.4 for x in histogram.keys()], [v / num_points for v in histogram.values()], 0.8, color='0.75')
    mu = p * n # Mean
    sigma = math.sqrt(n * p * (1 - p)) # Standard Deviation
    xs = range(min(data), max(data) + 1)
    ys = [normal_cdf(x + 0.5, mu, sigma) - normal_cdf(x - 0.5, mu, sigma) for x in xs] # One unit interval
    plt.plot(xs,ys) # Use a Line chart to show a normal aproximation
    plt.title("Binomial Distribution vs. Normal Distribution" )
    plt.show()

make_hist(0.75, 100, 10000)
```

```python
# Probability density function - Describes the relative probability of a random variable taking a given value;

def normal_pdf(x, mu=0, sigma=1):
    sqrt_two_pi = math.sqrt(2 * math.pi)
    return (math.exp(-(x-mu) ** 2 / 2 / sigma ** 2) / (sqrt_two_pi * sigma))

# Histogram

def make_hist(p, n, num_points):
    data = [binomial(n, p) for _ in range(num_points)]
    histogram = Counter(data)
    # Use a Bar chart to show the  sample current binomials
    plt.bar([x - 0.4 for x in histogram.keys()], [v / num_points for v in histogram.values()], 0.8, color='0.75')
    mu = p * n # Mean
    sigma = math.sqrt(n * p * (1 - p)) # Standard Deviation
    xs = range(min(data), max(data) + 1)
    ys = [normal_pdf(x, mu, sigma) for x in xs]
    plt.plot(xs,ys) # Use a Line chart to show a normal aproximation
    plt.title("Binomial Distribution vs. Normal Distribution")
    plt.show()

make_hist(0.75, 100, 10000)
```
